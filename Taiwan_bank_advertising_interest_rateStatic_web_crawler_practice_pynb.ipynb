```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

```python
url = 'https://rate.bot.com.tw/xrt?Lang=zh-TW'
response = requests.get(url)
data = response.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
datas = soup.find('table',class_='table table-striped table-bordered table-condensed table-hover')



```

```python
# 幣別
tbody = datas.find_all("tbody")
tr = tbody[0].find_all("tr")
currency = []
for i in tr:
  c = i.find('div', class_='visible-phone print_hide').get_text(strip=True)
  currency.append(c)
```

```python
# 現金本行買入
cash_buy = []
for i in tr:
  i = i.find_all('td', class_='rate-content-cash text-right print_hide')[0].get_text(strip=True)
  cash_buy.append(i)


```

```python
# 現金本行賣出
cash_sell = []
for i in tr:
  i = i.find_all('td', class_='rate-content-cash text-right print_hide')[1].get_text(strip=True)
  cash_sell.append(i)
```

```python
# 即期買入
spot_buy = []
for i in tr:
  i = i.find_all('td', class_='rate-content-sight text-right print_hide')[0].get_text(strip=True)
  spot_buy.append(i)
```

```python
# 即期賣出
spot_sell = []
for i in tr:
  i = i.find_all('td', class_='rate-content-sight text-right print_hide')[1].get_text(strip=True)
  spot_sell.append(i)
```

```python
# 創建表格
df = pd.DataFrame({'幣別':currency, '現金匯率本行買入':cash_buy, '現金匯率本行賣出':cash_sell, '即期匯率買入':spot_buy, '即期匯率賣出':spot_sell})
df
```
